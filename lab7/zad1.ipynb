```python
import numpy as np
from matplotlib import pyplot as plt

# tworzymy tablice o wymiarach 128x128x3 (3 kanaly to RGB)
# uzupelnioną zerami = kolor czarny
data = np.zeros((128, 128, 3), dtype=np.uint8)


# chcemy zeby obrazek byl czarnobialy,
# wiec wszystkie trzy kanaly rgb uzupelniamy tymi samymi liczbami
# napiszmy do tego funkcje
def draw(img, x, y, color):
    img[x, y] = [color, color, color]


# zamalowanie 4 pikseli w lewym górnym rogu
draw(data, 5, 5, 100)
draw(data, 6, 6, 100)
draw(data, 5, 6, 255)
draw(data, 6, 5, 255)

draw(data, 1, 1, 222)

# rysowanie kilku figur na obrazku
for i in range(128):
    for j in range(128):
        if (i-64)**2 + (j-64)**2 < 900:
            draw(data, i, j, 200)
        elif i > 100 and j > 100:
            draw(data, i, j, 255)
        elif (i-15)**2 + (j-110)**2 < 25:
            draw(data, i, j, 150)
        elif (i-15)**2 + (j-110)**2 == 25 or (i-15)**2 + (j-110)**2 == 26:
            draw(data, i, j, 255)

# konwersja macierzy na obrazek i wyświetlenie
print(data.shape)
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
kernel = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
```

```python
def multiply_minor(minor,kernel):
    result = np.uint8(0)
    for y in range(3):
        for x in range(3):
            # print(minor[y][x], ' * ',kernel[y][x],' = ', minor[y][x] * kernel[y][x], '+')
            result += minor[y][x] * kernel[y][x]
    return result
```

```python
def minorize(matrix):
    minors = []
    for i in range(matrix.shape[0] - 2):
        for j in range(matrix.shape[1] - 2):
            submatrix = matrix[i:i+3, j:j+3]
            minors.append(submatrix)
    return minors
```

```python
result = [[],[],[]]
for z in range(3):
    result_layer = []
    matrix = data[:,:,z]
    minors = []
    minors = minorize(matrix)
    for minor in minors:
        result_layer.append(multiply_minor(minor,kernel))
    result_layer = np.array(result_layer)
    chunks = result_layer.reshape(-1,126)
    result[z].append(chunks)
```

```python
result = np.transpose(np.squeeze(result),(1,2,0))
```

```python
plt.imshow(result, interpolation='nearest')
plt.show()
```

```python
import math
def convolute(image, kernel_x, kernel_y, stride=1):
    print(image.shape)
    image_height, image_width, img_depth = image.shape
    kernel_height, kernel_width = kernel_x.shape

    output_height = (image_height - kernel_height) // stride + 1
    output_width = (image_width - kernel_width) // stride + 1
    output = np. zeros((output_height, output_width), dtype=np.uint8)

    for i in range(output_height - 1):
        for j in range(output_width - 1):
            sum_x = np.sum(image[i*stride:i*stride+kernel_height, j*stride:j*stride+kernel_width]*kernel_x)
            sum_y = np.sum(image[i*stride:i*stride+kernel_height, j*stride:j*stride+kernel_width]*kernel_y)
            output[i + 1, j + 1] = math.sqrt(pow(sum_x, 2) + pow(sum_y, 2))
    return output
```

```python
vertical_component_kernel = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
horizontal_component_kernel = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

convoluted_data_1 = convolute(data, horizontal_component_kernel, vertical_component_kernel )
plt.imshow(convoluted_data_1, interpolation="nearest")
```
