```python
import pyswarms as ps
import numpy as np
import math
from pyswarms.utils.functions import single_obj as fx
from pyswarms.utils.plotters import plot_cost_history
import matplotlib.pyplot as plt
```

# Set the limits to [0, 1] for all dimensions

```python
bounds = (np.zeros(6), np.ones(6))
```

# Define the custom objective function for endurance

```python
def endurance(x):
  return -1 * (math.exp(-2*(x[1]-math.sin(x[0]))**2)+math.sin(x[2]*x[3])+math.cos(x[4]*x[5]))
```

# Define the function that will be used by the optimizer

```python
def f(x):
    # Initialize an empty array to hold the cost values for each particle
    cost = np.zeros(x.shape[0])
    # Loop over each particle and evaluate the cost
    for i in range(x.shape[0]):
        cost[i] = endurance(x[i])
    return cost
```

# Set the options for the optimizer

```python
options = {'c1': 0.5, 'c2': 0.3, 'w':0.7}
```

# Initialize the optimizer

```python
optimizer = ps.single.GlobalBestPSO(n_particles=10, dimensions=6, options=options, bounds=bounds)
```

# Run the optimization

```python
cost, pos = optimizer.optimize(f, iters=1000)
```

# Plot the cost history

```python
plot_cost_history(optimizer.cost_history)
plt.show()
```

# Print the best cost and position

```python
print("Best cost:", cost)
print("Best position:", pos)
```
